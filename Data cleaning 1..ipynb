```python
#import library for data cleaning 
import pandas as pd 
import numpy as np

# import data into python
Building_data = pd.read_csv("C:/Users/vikar/Desktop/Viraj Karkar/Technocolab/Building_Permits.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
#see first five records from the record
Building_data.head()
```

```python
# get the number of missing data points per column
missing_values_count = Building_data.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[0:15]
```

```python
# how many total missing values do we have?
total_cells = np.product(Building_data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
Building_data_with_na_dropped = Building_data.dropna(axis=1)
cols_in_original_dataset = Building_data.shape[1]
cols_in_na_dropped = Building_data_with_na_dropped.shape[1]
dropped_columns = cols_in_original_dataset - cols_in_na_dropped
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % Building_data.shape[1])
print("Columns in na's dropped: %d" % Building_data_with_na_dropped.shape[1])
```
